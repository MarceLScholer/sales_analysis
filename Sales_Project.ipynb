```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('sales_data.csv', sep=';')
df.head()
```

```python
print(df.columns.tolist())
```

```python
df = df.rename(columns={'Producto': 'Product'})
df = df.rename(columns={'Vendedor': 'Sales Rep'})
df = df.rename(columns={'Ventas ': 'Sales'})
df.head()
```

```python
sales_region = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

sales_region.plot(kind='bar', color=['#6B6BE9','#E96B6B','#6BE96C'], edgecolor='black')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales ($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
sales_product = df.groupby('Product')['Sales'].sum().sort_values(ascending=False)

sales_product.plot(kind='bar', color=['#6B6BE9','#E96B6B','#6BE96C'], edgecolor='black')
plt.title('Sales by Product')
plt.xlabel('Product')
plt.ylabel('Sales ($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
sales_salesrep = df.groupby('Sales Rep')['Sales'].sum().sort_values(ascending=False)

sales_salesrep.plot(kind='barh', color='#6B6BE9', edgecolor='black')
plt.title('Sales by Sales Rep')
plt.xlabel('Ventas ($)')
plt.tight_layout()
plt.show()
```
